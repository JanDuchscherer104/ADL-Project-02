```python
from ingredient_classifier import *
```

```python
# for model_type in ModelType:
cfg = ExperimentConfig(
    module_config=ImgClassifierParams(
        model=ModelType.VISION_TRANSFORMER,
        train_head_only=True,
    ),
    trainer_config=TrainerConfig(
        fast_dev_run=False,
        max_steps=-1,
        max_epochs=10,
    ),
    is_debug=False,
    datamodule_config=DatamoduleParams(
        train_ds=GroceryDatasetConfig(
            transforms_type=TransformsType.TRAIN_FROM_SCRATCH,
        ),
        batch_size=32,
    )
)
trainer, module, datamodule = cfg.setup_target()
trainer.fit(module, datamodule)
```

```python
# test dataloader, get a batch an plot grid using torchvision.utils.make_grid
from torchvision.utils import make_grid
import matplotlib.pyplot as plt

batch = next(iter(datamodule.train_dataloader()))
images = batch[0]
grid = make_grid(images, nrow=8)

# Plot grid of training images
plt.figure(figsize=(15,15))
plt.imshow(grid.permute(1,2,0).numpy())
plt.axis('off')
plt.show()
```
